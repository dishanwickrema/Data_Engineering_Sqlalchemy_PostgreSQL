```python
nbhd = Branch_General_Profile[['NBHD Number','NBHD Name']]
nbhd = nbhd.sort_values('NBHD Number')
print(len(nbhd))
nbhd.drop_duplicates(subset = 'NBHD Number', keep = 'first', inplace=True)
print(len(nbhd))
nbhd = nbhd.rename(columns={'NBHD Number':'id','NBHD Name':'name'})
nbhd = nbhd.reset_index(drop=True)
nbhd.head()
```

```python
ward_region = Branch_General_Profile[['Ward Number','Ward Region']]
ward_region = ward_region.sort_values('Ward Number')
print(len(ward_region))
ward_region.drop_duplicates(subset = 'Ward Number', keep = 'first', inplace=True)
print(len(ward_region))
ward_region = ward_region.rename(columns={'Ward Number':'id','Ward Region':'name'})
ward_region = ward_region.reset_index(drop=True)
ward_region.head()
```
